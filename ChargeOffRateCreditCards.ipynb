```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
reader = pd.read_csv("Resources/Default2.csv")
reader
```

```python
reader = reader[['date', 'value']].copy()
reader.head(20)
```

```python
reader = reader.rename(columns = {"date" : "Date", "value" : "Charge_off_Rate"})
```

```python
reader.drop(0)
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/CreditCardAnalysis2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
reader.to_sql(name='fredcor', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('SELECT * FROM fredcor', con=engine).head(300)
```
