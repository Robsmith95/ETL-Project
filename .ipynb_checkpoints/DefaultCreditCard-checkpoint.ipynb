```python
import pandas as pd
from sqlalchemy import create_engine
reader = pd.read_csv("Resources/Default.csv")
reader
```

```python
reader = reader[['X1', 'X2', 'X3', 'X5']].copy()
reader.head(20)
```

```python
reader = reader.rename(columns = {"X1" : "LIMIT_BAL", "X2" : "SEX", "X3" : "EDUCATION", "X5" : "AGE"})
```

```python
reader.drop(0)
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/CreditCardAnalysis"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
reader.to_sql(name='balance_limit', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('SELECT * FROM balance_limit', con=engine).head()
```
