```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
reader = pd.read_csv("Resources/Default2.csv")
reader
```

```python
reader = reader[['date', 'value']].copy()
reader.head(20)
```

```python
type(False)
```

```python
booleans = []
for length in reader.date:
    if length >= "2008-01-01" and length < "2010-12-31":
        booleans.append(True)
    else: 
        booleans.append(False)
```

```python
len(booleans)
```

```python
is_long = pd.Series(booleans)
is_long.head()
```

```python
reader[is_long]
```

```python
reader2 = reader[is_long]
reader2
```

```python
reader2 = reader2.rename(columns = {"date" : "Date", "value" : "Charge_off_Rate"})
reader2
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/CreditCardAnalysis2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
reader2.to_sql(name='fredcor08_10', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('SELECT * FROM fredcor08_10', con=engine).head(30)
```
